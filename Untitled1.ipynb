```python
# prices has MultiIndex columns with names=['Price','Ticker']
aapl = prices.xs('AAPL', level='Ticker', axis=1)
```

```python
aapl.head()
```

```python
aapl.isna().any()
```

```python
import pandas as pd

# 126 trading days ahead
forecast_days = 126

# Create a new column 'Future_Close' that is the Close price shifted up by 126 days
aapl['Future_Close'] = aapl['Close'].shift(-forecast_days)
```

```python
aapl.head()
```

```python

# Features
X = aapl[['Open', 'High', 'Low', 'Close', 'Volume']].iloc[:-forecast_days]

# Target
y = aapl['Future_Close'].iloc[:-forecast_days]
```

```python
split = int(len(X) * 0.8)  # 80% training, 20% test

X_train = X.iloc[:split]
X_test  = X.iloc[split:]

y_train = y.iloc[:split]
y_test  = y.iloc[split:]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Predict
y_pred = model.predict(X_test_scaled)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
import numpy as np

latest_data = aapl[['Open','High','Low','Close','Volume']].iloc[-1].values.reshape(1,-1)
latest_scaled = scaler.transform(latest_data)

future_price = model.predict(latest_scaled)
print("Predicted Apple price in 6 months:", future_price[0])
```
